Import all the packages

```python
import numpy as np
import pandas as pd
from scipy.io.arff import loadarff 
# arff is just another file format
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

Load the data

```python
phis_data = loadarff("PhishingData.arff")
phis_data
```

Convert data to numpy format

```python
phis_data_array = np.array(phis_data[0])
phis_data_array
```

Converting numpy array to pandas dataframe and converting columns to numeric type.

```python
phis_df = pd.DataFrame(phis_data_array).apply(pd.to_numeric)
```

```python
print(phis_df.shape)
```

```python
phis_df.head()
```

```python
phis_df.tail()
```

### Creating the label

We are trying to predict the 'Result' column in the data frame. We now create the label.

```python
label = phis_df['Result'].values
print(label)
```

```python
# phis_df['Result'].value_counts()
phis_df.groupby('Result').size()
```

According to original data there are 6157 proper websites and 4898 malicious websites.

### Plot to malicious and valid website

```python
sns.countplot('Result', data=phis_df)
plt.show()
```

### Creating the features or independant variables

```python
print(phis_df.columns[:-1])
```

```python
features = phis_df.columns[:-1]
predictors = phis_df[features].values
predictors
```

Correlation among the columns

```python
corr = phis_df.corr()
corr
```

### Splitting the data into training and testing data

```python
x = predictors
y = label

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.30, random_state=88)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

### Technique to choose the right no.of neighbours using cross validation.

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
list_k = [1,3,5,7,9,11,13,15] 
# We usually go with odd number of neighbours to avoid the tie.
# cv = 5 indicates are splitting 4 parts for building model and 1 part for validating model

for k in list_k:
    clf_knn = KNeighborsClassifier(n_neighbors = k)
    score_knn = cross_val_score(clf_knn, x_test, y_test, cv=5)
    print("K :", k)
    print("CV score :" + str(score_knn))
    print("CV mean score" + str(score_knn.mean()))
    print("")
```

Best CV mean score is obtained for k=11 , Hence we choose 11 neighbors for the model.

### Building model and training the model

```python
clf_knn = KNeighborsClassifier(n_neighbors = 11)
clf_knn = clf_knn.fit(x_train, y_train)
```

### Make prediction from the model

```python
y_pred = clf_knn.predict(x_test)
```

### Evaluate the model performance 

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

cr = classification_report(y_test, y_pred)
print(cr)
```

```python
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
sns.heatmap(cm, xticklabels = ['Actual NO', 'Actual YES'], yticklabels = ['Predicted NO', 'Predicted YES'], annot=True, fmt='d')
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(accuracy)
```
